Most recommendation models consist of building a user-by-item matrix with some sort of “interaction” number in each cell. If one includes the numerical ratings that users give items, then this is called an explicit feedback model. Alternatively, one may include implicit feedback which are actions by a user that signify a positive or negative preference for a given item (such as viewing the item online). These two scenarios often must be treated differently.

In the case of the MovieLens dataset, we have ratings, so we will focus on explicit feedback models

```python
movie_matrix = data.pivot_table(index='userId', columns='title', values='rating')
movie_matrix = movie_matrix.fillna(0)
movie_matrix
```

### Basic recommendation based on score correlation

```python
movie = 'First Blood (Rambo: First Blood) (1982)'

#find correlated movies
recommendations = movie_matrix.corrwith(movie_matrix[movie])

#sort them from high to low
recommendations = recommendations.sort_values(ascending=False)

#make the table look pretty and print the top 10
pd.DataFrame(recommendations, columns=['recommendation strength']).head(10)
```

### More complex recommendation based on k-nearest neighbour

```python
from scipy.sparse import csr_matrix

def create_X(df):
    """
    Generates a sparse matrix from ratings dataframe.
    
    Args:
        df: pandas dataframe
    
    Returns:
        X: sparse matrix
        user_mapper: dict that maps user id's to user indices
        user_inv_mapper: dict that maps user indices to user id's
        movie_mapper: dict that maps movie id's to movie indices
        movie_inv_mapper: dict that maps movie indices to movie id's
    """
    N = df['userId'].nunique()
    M = df['movieId'].nunique()

    user_mapper = dict(zip(np.unique(df["userId"]), list(range(N))))
    movie_mapper = dict(zip(np.unique(df["movieId"]), list(range(M))))
    
    user_inv_mapper = dict(zip(list(range(N)), np.unique(df["userId"])))
    movie_inv_mapper = dict(zip(list(range(M)), np.unique(df["movieId"])))
    
    user_index = [user_mapper[i] for i in df['userId']]
    movie_index = [movie_mapper[i] for i in df['movieId']]

    X = csr_matrix((df["rating"], (movie_index, user_index)), shape=(M, N))
    
    return X, user_mapper, movie_mapper, user_inv_mapper, movie_inv_mapper
```

```python
X, user_mapper, movie_mapper, user_inv_mapper, movie_inv_mapper = create_X(ratings)
```

```python
from sklearn.neighbors import NearestNeighbors

def find_similar_movies(movie_id, X, k, metric='cosine', show_distance=False):
    """
    Finds k-nearest neighbours for a given movie id.
    
    Args:
        movie_id: id of the movie of interest
        X: user-item utility matrix
        k: number of similar movies to retrieve
        metric: distance metric for kNN calculations
    
    Returns:
        list of k similar movie ID's
    """
    neighbour_ids = []
    
    movie_ind = movie_mapper[movie_id]
    movie_vec = X[movie_ind]
    k+=1
    kNN = NearestNeighbors(n_neighbors=k, algorithm="brute", metric=metric)
    kNN.fit(X)
    if isinstance(movie_vec, (np.ndarray)):
        movie_vec = movie_vec.reshape(1,-1)
    neighbour = kNN.kneighbors(movie_vec, return_distance=show_distance)
    for i in range(0,k):
        n = neighbour.item(i)
        neighbour_ids.append(movie_inv_mapper[n])
    neighbour_ids.pop(0)
    return neighbour_ids
```

```python
movie_titles = dict(zip(movies['movieId'], movies['title']))

movie_id = 1

similar_ids = find_similar_movies(movie_id, X, k=10)
movie_title = movie_titles[movie_id]

print(f"Because you watched {movie_title}")
for i in similar_ids:
    print(movie_titles[i])
```

### Content Filtering

```python
movies['genres'] = movies['genres'].str.split('|')
```

```python
#movies.index = movies['title']
movie_features = pd.get_dummies(movies['genres'].explode()).sum(level=0)
movie_features
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity = cosine_similarity(movie_features, movie_features)
cosine_similarity
```

```python
#Movie based on which you want to find other movies
movie = 'First Blood (Rambo: First Blood) (1982)'

#Finding the index number of that movie in the data
movie_index = dict(zip(movies['title'], list(movies.index)))
index = movie_index[movie]

#Creating similarity scores based on the movie
similarity_scores = list(enumerate(cosine_similarity[index]))

#Sorting the scores from most to least similar
similarity_scores = sorted(similarity_scores, key=lambda movie: movie[1], reverse=True)

#Getting the index numbers of those movies, and getting their titles based on that
similar_movies = [i[0] for i in similarity_scores]
pd.DataFrame(movies['title'].iloc[similar_movies]).head(10)
```
